### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

Tienda = pd.read_csv(url)
Tienda2 = pd.read_csv(url2)
Tienda3 = pd.read_csv(url3)
Tienda4 = pd.read_csv(url4)

print("Tienda")
display(tienda.head())
print("Tienda 2")
display(tienda2.head())
print("Tienda 3")
display(tienda3.head())
print("Tienda 4")
display(tienda4.head())

tienda.head()
```

```python
facturacion_Tienda = Tienda['Precio'].sum()
facturacion_Tienda2 = Tienda2['Precio'].sum()
facturacion_Tienda3 = Tienda3['Precio'].sum()
facturacion_Tienda4 = Tienda4['Precio'].sum()


print(f"Facturación Tienda 1: ${facturacion_Tienda:,.2f}")
print(f"Facturación Tienda 2: ${facturacion_Tienda2:,.2f}")
print(f"Facturación Tienda 3: ${facturacion_Tienda3:,.2f}")
print(f"Facturación Tienda 4: ${facturacion_Tienda4:,.2f}")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
facturacion_df = pd.DataFrame({
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Facturación': [facturacion_Tienda, facturacion_Tienda2, facturacion_Tienda3, facturacion_Tienda4]
})

plt.figure(figsize=(8,5))
sns.barplot(x='Tienda', y='Facturación', data=facturacion_df, palette='viridis')
plt.title("Facturación Total por Tienda")
plt.ylabel("Facturación ($)")
plt.show()
```

# 2. Ventas por categoría

```python
# Contar ventas por categoría para cada tienda
ventas_categoria_tienda1 = Tienda['Categoría del Producto'].value_counts()
ventas_categoria_tienda2 = Tienda2['Categoría del Producto'].value_counts()
ventas_categoria_tienda3 = Tienda3['Categoría del Producto'].value_counts()
ventas_categoria_tienda4 = Tienda4['Categoría del Producto'].value_counts()

# Mostrar resultados
print("Ventas por Categoría del Producto - Tienda 1")
display(ventas_categoria_tienda1)
print("Ventas por Categoría del Producto - Tienda 2")
display(ventas_categoria_tienda2)
print("Ventas por Categoría del Producto - Tienda 3")
display(ventas_categoria_tienda3)
print("Ventas por Categoría del Producto - Tienda 4")
display(ventas_categoria_tienda4)
```

```python
# Renombrar columna en todas las tiendas
Tienda = Tienda.rename(columns={'Categoría del Producto': 'Categoría'})
Tienda2 = Tienda2.rename(columns={'Categoría del Producto': 'Categoría'})
Tienda3 = Tienda3.rename(columns={'Categoría del Producto': 'Categoría'})
Tienda4 = Tienda4.rename(columns={'Categoría del Producto': 'Categoría'})

ventas_categoria = Tienda['Categoría'].value_counts()

```

```python
# Lista de tiendas y nombres
tiendas = [Tienda, Tienda2, Tienda3, Tienda4]
nombres_tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

# Analizar ventas por categoría y graficar para cada tienda
for i, tienda in enumerate(tiendas):
    ventas_categoria = tienda['Categoría'].value_counts()

    print(f"Ventas por categoría - {nombres_tiendas[i]}")
    display(ventas_categoria)

    plt.figure(figsize=(8,5))
    sns.barplot(x=ventas_categoria.index, y=ventas_categoria.values, palette='coolwarm')
    plt.title(f"Ventas por Categoría - {nombres_tiendas[i]}")
    plt.ylabel("Número de Ventas")
    plt.xlabel("Categoría")
    plt.xticks(rotation=45)
    plt.show()
```

```python
# Gráfico comparativo
df_list = []

for i, tienda in enumerate(tiendas):
    temp_df = tienda['Categoría'].value_counts().reset_index()
    temp_df.columns = ['Categoría', 'Ventas']
    temp_df['Tienda'] = nombres_tiendas[i]
    df_list.append(temp_df)

ventas_comparativas = pd.concat(df_list)

# Gráfico comparativo con seaborn
plt.figure(figsize=(12,6))
sns.barplot(x='Categoría', y='Ventas', hue='Tienda', data=ventas_comparativas, palette='Set2')
plt.title("Comparativa de Ventas por Categoría entre Tiendas")
plt.ylabel("Número de Ventas")
plt.xlabel("Categoría")
plt.xticks(rotation=45)
plt.legend(title='Tienda')
plt.show()

```

# 3. Calificación promedio de la tienda


```python
# Calcular calificación promedio por tienda
for i, tienda in enumerate(tiendas):
    calificacion_promedio = tienda['Calificación'].mean()
    print(f"Calificación promedio - {nombres_tiendas[i]}: {calificacion_promedio:.2f}")

# Crear DataFrame para graficar
calificaciones_df = pd.DataFrame({
    'Tienda': nombres_tiendas,
    'Calificación Promedio': [tienda['Calificación'].mean() for tienda in tiendas]
})

# Gráfico de barras
plt.figure(figsize=(8,5))
sns.barplot(x='Tienda', y='Calificación Promedio', data=calificaciones_df, palette='pastel')
plt.title("Calificación Promedio por Tienda")
plt.ylabel("Calificación")
plt.ylim(0,5)  # Asumiendo escala de 0 a 5
plt.show()
```

```python
# Calcular costo promedio de envío usando el nombre correcto
for i, tienda in enumerate(tiendas):
    costo_envio_promedio = tienda['Costo de envío'].mean()
    print(f"Costo de envío promedio - {nombres_tiendas[i]}: ${costo_envio_promedio:.2f}")

# Crear gráfico comparativo
envios_df = pd.DataFrame({
    'Tienda': nombres_tiendas,
    'Costo Promedio Envío': [tienda['Costo de envío'].mean() for tienda in tiendas]
})

plt.figure(figsize=(8,5))
sns.barplot(x='Tienda', y='Costo Promedio Envío', data=envios_df, palette='magma')
plt.title("Costo Promedio de Envío por Tienda")
plt.ylabel("Costo Promedio ($)")
plt.show()
```

# 4. Productos más y menos vendidos

```python
# Crear resumen comparativo
resumen_df = pd.DataFrame({
    'Tienda': nombres_tiendas,
    'Facturación Total': [tienda['Precio'].sum() for tienda in tiendas],
    'Calificación Promedio': [tienda['Calificación'].mean() for tienda in tiendas],
    'Costo Promedio Envío': [tienda['Costo de envío'].mean() for tienda in tiendas]
})

# Imprimir resumen
display(resumen_df)
```

```python
# Grafico comparativo de Facturación
plt.figure(figsize=(8,5))
sns.barplot(x='Tienda', y='Facturación Total', data=resumen_df, palette='viridis')
plt.title("Facturación Total por Tienda")
plt.ylabel("Facturación ($)")
plt.show()

# Grafico comparativo de Calificación
plt.figure(figsize=(8,5))
sns.barplot(x='Tienda', y='Calificación Promedio', data=resumen_df, palette='pastel')
plt.title("Calificación Promedio por Tienda")
plt.ylabel("Calificación")
plt.ylim(0,5)
plt.show()

# Grafico comparativo de Costo de Envío
plt.figure(figsize=(8,5))
sns.barplot(x='Tienda', y='Costo Promedio Envío', data=resumen_df, palette='magma')
plt.title("Costo Promedio de Envío por Tienda")
plt.ylabel("Costo Promedio ($)")
plt.show()
```

# 5. Envío promedio por tienda

```python
# Promedio de envíos por tienda (suponiendo que cada fila es un envío)
for i, tienda in enumerate(tiendas):

   # total de filas es el total de envíos

    promedio_envios = len(tienda)  # total de filas es el total de envíos
    print(f"Total de envíos - {nombres_tiendas[i]}: {promedio_envios}")
```

```python
# Graficar total de envios
total_envios = [len(tienda) for tienda in tiendas]

envios_df = pd.DataFrame({
    'Tienda': nombres_tiendas,
    'Total de Envíos': total_envios
})

display(envios_df)

plt.figure(figsize=(8,5))
sns.barplot(x='Tienda', y='Total de Envíos', data=envios_df, palette='coolwarm')
plt.title("Total de Envíos por Tienda")
plt.ylabel("Número de Envíos")
plt.show()
```

#### **Recomendacion Final** ####


```python
# Crear resumen final
resumen_final = pd.DataFrame({
    'Tienda': nombres_tiendas,
    'Facturación Total': [tienda['Precio'].sum() for tienda in tiendas],
    'Calificación Promedio': [tienda['Calificación'].mean() for tienda in tiendas],
    'Costo Promedio Envío': [tienda['Costo de envío'].mean() for tienda in tiendas],
    'Total de Envíos': [len(tienda) for tienda in tiendas]
})


display(resumen_final)
```

```python
# Crear un score de eficiencia (menor facturacion + menor calificación + mayor costo de envío)
resumen_final['Score Ineficiencia'] = (
    resumen_final['Facturación Total'].rank(ascending=True) +   # menor facturacion = peor
    resumen_final['Calificación Promedio'].rank(ascending=True) +  # menor Calificacion = peor
    resumen_final['Costo Promedio Envío'].rank(ascending=False)    # mayor costo de envío = peor
)

# Ordenar por Score Ineficiencia
resumen_final = resumen_final.sort_values(by='Score Ineficiencia', ascending=False)

# Mostrar tienda menos eficiente
tienda_menor_eficiencia = resumen_final.iloc[0]
print(f"Tienda menos eficiente: {tienda_menor_eficiencia['Tienda']}")
```
